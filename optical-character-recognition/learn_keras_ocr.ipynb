```python
import matplotlib.pyplot as plt
import keras_ocr
import cv2 as cv
import os
from google.colab.patches import cv2_imshow # for image display
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
img = plt.imread('/content/images/1.png')
plt.imshow(img)
```

```python
images = []
for img_path in os.listdir('/content/images'):
    try:
        img = plt.imread(os.path.join('/content/images', img_path))
        images.append(img)
    except:
        continue

print('Number of images', len(images))
```

```python
fig, axes = plt.subplots(ncols=5, figsize=(25, 5))
for img, ax in zip(images, axes):
    ax.imshow(img)
    # Hide grid lines
    ax.grid(False)

    # Hide axes ticks
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
img_canny = cv.Canny(img, 100, 200, 3)
cv2_imshow(img_canny)
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```
